Here is the function which takes a one-dimensional Numpy array of data and returns x and y values for plotting the ECDF.

```python
def ecdf(data):
    sortedData = np.sort(data)
    n = sortedData.size
    yValues = []
    for i in range(n):
        yValues.append((i + 1) / n)
    return (list(sortedData), yValues)
```

<div class="alert alert-info"> **Always include docstring with your functions!** This is to future proof yourself as well as make it easier for the TAs to understand what your function is doing. While your code works as written, the same steps can be performed in a simpler manner just by returning `x,y = np.sort(data), np.arange(1, len(data) + 1) / len(data)`. </div>

Here we separate the data into two Numpy arrays.

```python
labeled_array = np.array(df["labeled"])
unlabeled_array = np.array(df["unlabeled"])
```

Here ecdf() is used to replicate the ECDF in Fig. 2a of the Gardner, Zanic, et al. paper.

```python
(xL, yL) = ecdf(labeled_array)
(xUL, yUL) = ecdf(unlabeled_array)

line1 = plt.plot(xL, yL, marker='.', linestyle='none', label='Labeled', color='g')
line2 = plt.plot(xUL, yUL, marker='.', linestyle='none', label='Unlabeled', color='b')
plt.xlabel('Catastrophe time, s')
plt.ylabel('Cumulative distribution')
plt.legend(loc='lower right')
```

1.3 (e) 

Cumulative distribution function (CDF) plot shows how each element relates to the sum of all the elements, since the difference
between two successive cumulative sums is portrayed in relation to the cumulative total. For example, if each element is very 
small compared to the cumulative sum, then the CDF plot lacks abrupt changes and appears smooth and continuous. However, if 
at least one element is comparable to the sum, the CDF plot looks steep and bumpy. Notice that CDF isn't very conducive
to analyzing how different elements relate to one another - each bin is considered in the context of the sum of all the bins. 
Consequently, it's easier to look at various percentiles using a CDF plot. 

Regular histogram is useful for showing how different elements relate to one another since the height of each bar represents 
the corresponding frequency of interest and the difference between two heights helps relate the two bins in the set. However, 
using a regular histogram makes it difficult to relate each element to the sum of all the elements. It is not immediately obvious from a regular histogram what percentage of the sum each bin corresponds to. In the limit as each bin gets smaller, the sum of all the bins gets harder to estimate just by looking at the graph. Therefore, a regular histogram isn't preferable for a holistic analysis of data with a focus on the sum.

1.3 (f)

ecdf_conventional() creates a 'continuous' plot by creating additional x and y values at small intervals relative to the distance between the data points they are between. 

```python
def ecdf_conventional(data):
    sData = np.sort(data)
    size = sData.size
    returnX = []
    returnY = []
    count = 1
    for i in range(size):
        if i == size - 1:
            end = sData[i] + (sData[i] - sData[i - 1])
        else:
            interval = (sData[i + 1] - sData[i]) / 100
            end = sData[i+1]
        temp = sData[i]
        
        while (temp < end - 0.0001):
        # The 0.0001 is the epsilon to account for adding duplicate x values due to floating error 
            returnX.append(temp)
            returnY.append(count/size)
            temp += interval
        count += 1
    
    return (returnX, returnY)
```

```python
(xL, yL) = ecdf_conventional(labeled_array)
(xUL, yUL) = ecdf_conventional(unlabeled_array)

line1 = plt.plot(xL, yL, marker='.', linestyle='none', label='Labeled', color='g')
line2 = plt.plot(xUL, yUL, marker='.', linestyle='none', label='Unlabeled', color='b')
plt.xlabel('Catastrophe time, s')
plt.ylabel('Cumulative distribution')
plt.legend(loc='lower right')
```
