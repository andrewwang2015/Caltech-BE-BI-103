```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read data file and print
f = '../data/han_wang_gal4_temperature.xlsx'
df = pd.read_excel(f)
df
```

<div class="alert alert-info"><center> Good. 1/1</center></div>

```python
#We first melt so that we have species and temps as one column and the numerical values as the other
df_gt = pd.melt(df, var_name='Species and temps', value_name='Values')
df_gt
#speciesTemps is the first column of the species and temps
speciesTemps = df_gt['Species and temps'].tolist()
species = []
temps = []
#values is the numerical data
values = df_gt['Values'].tolist()
#This for loop below parses data in the 'species and temps' column. For example if the data is 'sc15', it will append 'sc' to
#species and '15' to temps
for j in speciesTemps:
    species.append(j[:2])
    temps.append(j[2:])
df_gt
```

```python
#Create new column and import the numerical data into this column
df_gt["Fluorescence"] = values
#The "values" column was the second column, so we replace that with the list of temperatures
df_gt["Values"] = temps
#Make first column the species
df_gt["Species and temps"] = species
#Rename the three columns for final table which should be tidy
df_gt.columns = ['Gene Expression System', 'Temperature ($^\circ$C)', 'Fluorescence']
df_gt
```

```python
sns.swarmplot(x="Temperature ($^\circ$C)", y='Fluorescence', hue='Gene Expression System', data=df_gt, split=True)
```

We decided to use a swarmplot to plot all the data because each gene expression system had few enough points to be able to plot each individual datapoint without clustering the graph too much to the point where variability of the data could not be seen. This swarmplot with the split parameter allows us to compare gene expression systems at each temperature side by side which does not compromise the aesthetics that would be lost when the three gene expression systems are strictly overlaid over one another. 



Here are some alternative ways to present the data:

```python
sns.boxplot(x="Temperature ($^\circ$C)", y='Fluorescence', hue='Gene Expression System',data=df_gt, palette="PRGn")
```

The box and whisker plot allows us to focus on key parameters in each distribution. For example, the 25th, 50th and 75th percentiles can be seen from the edges of each box. In addition, whiskers make it easy to identify outliers that extend beyond 1.5 x IQR  and draw a useful boundary for the data set. Notice that the boundaries are marked by the ends of the whiskers and the outliers are plotted as individual points beyond the corresponding whiskers. 

Even though the box and whisker plot helps us glean some insight into the distributions, we don't think it's the best way to represent our data. First of all, the assumptions about the "core" of the distribution (25th and 75th percentiles) are arbitrary and potentially misleading - since we don't see the distribution within each box, the only visual data we have to analyze the "core" are the endpoints and the midpoint. For any box-and-whisker plot, there exist infinitely many unique distributions with largely varying properties. 

Another reason why we prefer the swarm plot is that none of the data are omitted. While the box and whisker plot distills the data into a manageable and visually intuitive form, it takes away a lot of information about the location and relative importance of each individual point. The swarm plot displays the data as given, without any visual aids, but it preserves the uniqueness of the data. For each swarm plot, there can only exist one data set (Not counting any repeated points that get stacked on top of one another). 




```python
sns.set(style="whitegrid", palette="pastel", color_codes=True)
sns.violinplot(x="Temperature ($^\circ$C)", y='Fluorescence', hue='Gene Expression System', data=df_gt, inner="quart")
```

The violin plot simultaneously displays many interesting facts about the data. It allows us to compare the relative variance of each distribution, given by the length of each violin string, and the number of data points that fall within a certain range, given by the width of each string. Notice that the outliers in m3 at 15 $^\circ$C extend the string outward, emphasizing the outliers. Similarly, the strings corresponding to sk at 15 and 20 $^\circ$C are wider than their sc counterparts at higher fluorescence values, indicating that the median fluorescence intensity of sk exceeds that of sc. 

While the violin plot can be useful to analyze, we find it less preferable than the swarm plot for two main reasons: 

1. It adds artificial data that do not exist in the original set in order to present a smooth and pleasant image of each distribution. Notice that the bottom outlier in the 15 $^\circ$C m3 set and the two lowest temperature sc sets fall below the zero point. Considering that negative fluorescence doesn't make any physical sense, this artificial representation can be misleading. It is best the preserve the properties of the original data.

2. It bins the data in an arbitrary fashion, biasing our analysis. While the binnings may make sense, there is no guarantee that they will for every data set. Considering that unimodal distributions may appear bimodal when binned incorrectly, it is important to avoid arbitrary binning when possible. The swarmplot successfully displays the original data without adding any artificialities, so we prefer that over the violin plot as well.
