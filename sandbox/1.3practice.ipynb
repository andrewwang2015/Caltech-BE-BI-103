```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('../data/hw1data.csv', comment='#')
df
```

```python
df = df.dropna()
df
```

```python
'''
We first take a look at how the normed parameter affects the plots:
'''

#normed = false
df1 = df.rename(columns={'time to catastrophe with labeled tubulin (s)': 'labeled'})
_ = plt.hist(df1.labeled, bins=10, normed=False)
plt.xlabel('time to catastrophe with labeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
#normed = false

'''
It appears that if normed is true, the counts or number of observations become proportions so that
when all added up, form a probability density (integral of histogram = 1)
'''
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=True)
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('Frequency');
```

```python
'''
Next,we take a look at the histtype
'''

#We first look at the 'step' option. The step option generates a line plot, unfilled
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=False, histtype='step')
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
#The stepfilled option generates a line plot, filled
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=False, histtype='stepfilled')
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
#For histtype = 'bar' (default), generates the usual bar-type histogram
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=False, histtype='bar')
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
#For histtype = 'barstacked,' if there was multiple data, it would be stacked.
#In this case, we only have data from one column, so barstacked and bar have no difference
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=False, histtype='barstacked')
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
'''
Now, looking at the 'bins' parameter. It determines how many "bars" or segments you will have
and specifically how many edges (bins + 1) will be outputted.
'''

#bins = 10
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=False)
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
#bins = 15
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=15, normed=False)
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
#bins = 20
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=20, normed=False)
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

```python
df = df.rename(columns={'time to catastrophe with unlabeled tubulin (s)': 'unlabeled'})
_ = plt.hist(df.unlabeled, bins=10, normed=False)
plt.xlabel('time to catastrophe with unlabeled tubulin (s)')
plt.ylabel('number of observations');
```

Official Histogram for Unlabeled

This is the best option for the unlabeled tubulin. We chose the normed option to be false because we are interested in the actual number of observations, not the proportion of it relative to the total. Playing around with the number of bins, it seems that 10 is the optimal number of bins to have so that the data is not supercongested into too few bins or super spread out if there are two many bins (this will cause bins that have just one observation to be of equal height). We also went with the default histtype because step and stepfilled options are less aesthetically pleasing and the barstacked is not necessary here as we are only taking one set of data.
