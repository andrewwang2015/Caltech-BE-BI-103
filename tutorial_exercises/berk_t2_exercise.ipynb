1 (b) One of the reasons why the data frame is not tidy is that each variable is described using two rows - one for the species name and the other for the variable in question. If the two rows were reduced to one, the data would have looked a lot tidier and easier to work with. Moreover, the 12 columns can be reduced to 3, as recommended in part (c) of the problem. 

For example, instead of presenting the data in 12 columns and 2 rows, we can use three columns and a single header, making it more convenient to use standard python data analysis tools. (Each point in the data set can be represented in three parameters: species, variable in question and the corresponding number in centimeters). 

Another reason why the data frame is not tidy is that different measurements are presented in the same row, making it difficult to disentangle each observation from the rest. For example, each row contains 12 measurements, which should each have its own row for a tidy data arrangement.

1 (c) This dataframe is tidy because the raw data is reduced to a form that is much easier to work with. Notice that it satisties all three conditions for "tidiness":

1. Each variable forms a column. (Indeed, each data point has three different (varying) defining parameters that make it unique, namely its species, measured parameter and the corresponding value). 
2. Each observation has its own row. 
3. Each element in the table belongs to the same experimental setup so that no data point falls out of place.

```python
df_new = df.stack(level=0)
```

```python
df_new
```

2(a).1: 

Here, the stack() command reorganized the data frame based on the uppermost level (which is level 0). In other words, it took the uppermost row of the data frame (which contained the three species) and reshuffled the data points so that the three species appeared as the rows of the resulting data frame. 

Alternatively, if level = 1 was chosen, the new rows would have been picked from the four "quantity" arguments - petal length, petal width, sepal length and sepal width. stack() switches the rows and columns of the data frame based on the level argument it receives. For example, see the shuffled data frame with level = 1 below:


```python
df_new_try = df.stack(level=1)
```

```python
df_new_try
```

As I mentioned earlier, when level = 1, the data frame is reorganized so that the four "quantity" arguments become the rows.

```python
df_new = df_new.sortlevel(1)
```

```python
df_new
```

Notice that sortlevel() sorted the species column in the order at which setosa, versicolor and virginica appeared earlier. If 0 was passed as an argument to sortlevel() instead, the sorting would not have taken place since species corresponds to level = 1 in the data frame.

```python
df_new = df_new.reset_index(level=1)
```

```python
df_new
```

Here, reset_index() quite intuitively reset the index of all the species, making them a part of the data. The argument level = 1 corresponded to the species column. In the resulting data frame, species names are no longer a part of the categories - they are merged with the numerical data, as the formatting suggests. In other words, reset_index() picked level = 1 (which corresponded to the species category) and removed the categorical indexing from the column, reducing the species category to data.

```python
df_new = df_new.rename(columns={'level_1': 'species'})
```

```python
df_new
```

As the name of the command suggests, rename() renamed the header named 'level_1' to 'species'.

2 (b). Yes, the resulting data frame is "tidy". It satisfies the three conditions for tidiness:

1. Each variable forms a column. For each plant, species, petal length, petal width, sepal length and sepal width vary and each of the five variables is given its own column.
2. Each observation is a separate row. Notice that each row corresponds to a separate plant and vice versa. Therefore, each observation (plant) is given its own row, as desired.
3. Each observational unit forms a separate table. The table above consists of data from the same experiment. Each data point shares the variables and the experimental context with every other data point in the set. Thus, the data set forms a coherent whole.
